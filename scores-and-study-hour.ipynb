```python
import matplotlib.pyplot as plt

X = df2['Hours']
y = df2['Scores']

# Create a scatter plot
plt.scatter(X, y, label='Scatter Plot', color='blue', marker='o')

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Scatter Plot')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y , test_size = .2,random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Assuming x_train is a Pandas Series
x_train_reshaped = np.array(x_train).reshape(-1, 1)

# Create a Linear Regression model
lr = LinearRegression()

# Fit the model with the reshaped data
lr.fit(x_train_reshaped, y_train)
```

```python
lr.predict(np.array(x_test).reshape(-1,1))
```

```python
y_test
```

```python
plt.scatter(df2['Hours'], df2['Scores'])

# Plot the regression line
x_line = np.linspace(min(df2['Hours']), max(df2['Hours']), 100).reshape(-1, 1)
y_line = lr.predict(x_line)
plt.plot(x_line, y_line, color='red', label='Regression Line')

plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.show()
```
